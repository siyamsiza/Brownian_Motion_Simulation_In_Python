## Random Walk

A **random walk** is a mathematical formalization of a path consisting of a succession of random steps. 

In one dimension, it can be thought of as a person starting at the origin and taking steps either to the left or right with equal probability.

## Random Walk Simulation 

```python
#Import the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Simulating random walks in python
#Number of steps
num_steps = 5000

#setting random steps (-1 == left) and (1 == right)
steps1 = np.random.choice([-1, 1], size=num_steps)
steps2 = np.random.choice([-1, 1], size=num_steps)
steps3 = np.random.choice([-1, 1], size=num_steps)

#cumulative sum to get the position of each step
walk1 = np.cumsum(steps1)
walk2 = np.cumsum(steps2)
walk3 = np.cumsum(steps3)

#Visualizing random walks
plt.figure(figsize=(15, 7))
plt.style.use('classic')
sns.lineplot(walk1)
sns.lineplot(walk2)
sns.lineplot(walk3)
plt.xlabel("Number of Steps")
plt.ylabel("Steps")
plt.title("Random Walks Simulation")
plt.grid(True)
plt.show()
```

## Brownian Motion 

**Brownian motion**, also known as a Wiener process, is a **continuous-time stochastic process** that describes random motion. 

It can be modeled similarly to a random walk but involves continuous movement rather than discrete steps.

## Brownian Motion Simulation

```python
#Brownian motion simulation

#parameters
num_steps = 5000
dt1 = 0.010
dt2 = 0.020
dt3 = 0.030

#Normal distribution random numbers
steps1 = np.random.normal(loc=0, scale=np.sqrt(dt1), size=num_steps)
steps2 = np.random.normal(loc=0, scale=np.sqrt(dt2), size=num_steps)
steps3 = np.random.normal(loc=0, scale=np.sqrt(dt3), size=num_steps)

#Cumulative sum to find the position at each time step
brownian_motion1 = np.cumsum(steps1)
brownian_motion2 = np.cumsum(steps2)
brownian_motion3 = np.cumsum(steps3)

#Plot the brownian motion
plt.figure(figsize=(15, 7))
sns.lineplot(brownian_motion1, color='g')
sns.lineplot(brownian_motion2, color='r')
sns.lineplot(brownian_motion3, color='b')
plt.xlabel('Number of Steps')
plt.ylabel('Position')
plt.title('Brownian Motion Simulation')
plt.grid(True)
plt.show()
```

### Applications:

Random walks are often used in various fields such as economics (stock prices), physics (particle diffusion), and computer science (algorithms).
Brownian motion is crucial in finance for modeling stock prices through stochastic calculus (e.g., Black-Scholes model).

### Compare and Contrast

In summary, while both simulations exhibit randomness and have similarities in their underlying principles, they differ significantly in terms of their mathematical formulation and applications.

The comparison between Random Walks and Brownian Motion reveals that while both exhibit randomness:
- Random Walks consist of discrete steps taken at fixed intervals whereas
- Brownian Motion represents continuous movement
- Each has distinct statistical properties and applications across various fields.
